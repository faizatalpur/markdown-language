#**what is markdown language?**

###it is lightweight,easy-to-read markup language used to format plain text with symbols to create headers,bolding,list and links which converted into HTML.

#**syntax**
###**header**:use # for levels or heading lines.
###**emphasis or bolding**: use * for italic and ** for bold the text.
###**lists**:use hyphens - or aesteriks * for bullet points or unordered list.
####example(use hyphens or *+space+text= listout or bullet points)
#####my favorite fruit list
- apple
- cherry
- pineapple
* orange
* mango
###**links**: use link text in[] brackets then URL in () brackets.
